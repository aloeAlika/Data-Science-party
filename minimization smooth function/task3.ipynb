```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

#f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

def func(x):
    return np.sin(x/5.)*np.exp(x/10.)+5*np.exp((-1)*x/2.)

x0 = 5.

#Попробуйте найти минимум, используя стандартные параметры в функции scipy.optimize.minimize 
#(т.е. задав только функцию и начальное приближение). 

a = minimize(func, x0).x
print(a)

x = np.linspace(1.,30., num = 100)
plt.plot(x, func(x), 'r')
plt.scatter(a, func(a))
plt.show()
```

```python
#Попробуйте менять начальное приближение и изучить, меняется ли результат. 

x0 = 15.

a = minimize(func, x0).x
print(a)

plt.plot(x, func(x), 'r')
plt.scatter(a, func(a))
plt.show()
```

```python
#Укажите в scipy.optimize.minimize в качестве метода BFGS 
#(один из самых точных в большинстве случаев градиентных методов оптимизации), 
#запустите из начального приближения x=2. 
#Градиент функции при этом указывать не нужно – он будет оценен численно.

x0 = 2.

m = minimize(func, x0, method='BFGS')
a = m.x
q = m.fun
print(m)

plt.plot(x, func(x), 'r')
plt.scatter(a, func(a))
plt.show()
```

```python
q1 = round(q, 2)
print(q1)
```

```python
x0 = 30.

m = minimize(func, x0, method='BFGS')
a = m.x
q = m.fun
print(m)

plt.plot(x, func(x), 'r')
plt.scatter(a, func(a))
plt.show()
```

```python
q2 = round(q, 2)
print(q2)

f = open('answer.txt', 'w')
f.write(str(q1)+ ' ' + str(q2))
f.close()

#В самом деле, градиентные методы обычно не решают задачу глобальной оптимизации, 
#поэтому результаты работы ожидаемые и вполне корректные.
```
