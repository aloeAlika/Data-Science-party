```python
import numpy as np
from scipy.spatial import distance
import re
import math


f = open('sentences.txt', 'r')
content = f.readlines()
f.close() 

s = []
words = {}
q = 0


#Каждая строка в файле соответствует одному предложению. 
#Считайте их, приведите каждую к нижнему регистру
#Произведите токенизацию, то есть разбиение текстов на слова.
#Не забудьте удалить пустые слова после разделения.


#Составьте список всех слов, встречающихся в предложениях. 
#Сопоставьте каждому слову индекс от нуля до (d - 1), где d — число различных слов в предложениях.

for line in content:
    temp = list(filter(lambda x : x != '', re.split('[^a-z]', line.lower())))
    s.append(temp)
    for word in temp:
        if word not in words:
            words[word] = q
            q += 1


n = len(content)
d = len(words)

#Создайте матрицу размера n * d, где n — число предложений. 
#Заполните ее: элемент с индексом (i, j) в этой матрице должен быть равен количеству вхождений j-го слова в i-е предложение.

matrix = np.zeros((n,d))

q = 0
for list_of_words in s:
    for word in list_of_words:
        matrix [q, words[word]] += 1
    q += 1

#Найдите косинусное расстояние от предложения в самой первой строке до всех остальных

list_of_distances = [0.] * n
list_of_distances[0] = float("inf")

for i in range(1,n):
    list_of_distances[i] = distance.cosine(matrix[0,:], matrix[i,:])

#Запишите номера двух предложений, ближайших к первому по косинусному расстоянию, в файл.

f = open('submission-1.txt', 'w')
q = list_of_distances.index(min(list_of_distances))
f.write(str(q) + ' ')
list_of_distances[q] = float("inf")
q = list_of_distances.index(min(list_of_distances))
f.write(str(q))
f.close()
```
