```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt

#f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

def func(x):
    return np.sin(x/5.)*np.exp(x/10.)+5*np.exp((-1)*x/2.)


def func2(x,w0,w1):
    return w0+w1*x

#Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A и свободный вектор b) 
#для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15. 

A = np.array([[1.,1.],
              [1.,15.]])

b = np.array([[func(1.)],[func(15.)]])

#Решите данную систему с помощью функции scipy.linalg.solve. 
w = linalg.solve(A, b)


#Нарисуйте функцию f и полученный многочлен. Хорошо ли он приближает исходную функцию?

x = np.linspace(1.,15., num = 100)
plt.plot(x, func(x), 'r')
plt.plot(x, func2(x,w[0],w[1]), 'g')
plt.show()
```

```python
def func3(x,w0,w1,w2):
    return w0+w1*x+w2*x**2

#Повторите те же шаги для многочлена второй степени, который совпадает с функцией f в точках 1, 8 и 15. 

A = np.array([[1.,1., 1.],
              [1.,8.,64.],
              [1.,15., 225.]])

b = np.array([[func(1.)],[func(8.)],[func(15.)]])

w = linalg.solve(A, b)

#Улучшилось ли качество аппроксимации?

plt.plot(x, func(x), 'r')
plt.plot(x, func3(x,w[0],w[1],w[2]), 'b')
plt.show()
```

```python
def func4(x,w0,w1,w2,w3):
    return w0+w1*x+w2*x**2+w3*x**3

#Повторите те же шаги для многочлена третьей степени, который совпадает с функцией f в точках 1, 4, 10 и 15. 

A = np.array([[1.,1., 1.,1.],
              [1.,4.,16.,64.],
              [1.,10.,100.,1000],
              [1.,15., 225.,3375.]])

b = np.array([[func(1.)],[func(4.)],[func(10.)],[func(15.)]])

w = linalg.solve(A, b)

plt.plot(x, func(x), 'r')
plt.plot(x, func4(x,w[0],w[1],w[2],w[3]), 'g')
plt.show()

#Хорошо ли он аппроксимирует функцию? 
```
